## Required packages
We will use three packages that you need to install separately in your jupyter/conda environment:
* <a href="https://scitools.org.uk/cartopy/docs/latest/" target="_blank">Cartopy</a>: a Python package designed for geospatial data processing in order to produce maps and other geospatial data analyses.
* <a href="" target="_blank">Matplotlib-scalbar</a>: a Python package designed for adding scalebars to figures and maps.
* <a href="https://osmnx.readthedocs.io/en/stable/#" target="_blank">OSMnx</a> : a Python package designed to download spatial geometries and model, project, visualize, and analyze street networks and other spatial data from OpenStreetMapâ€™s APIs.
* <a href="https://scikit-learn.org/stable/" target="_blank">scikit-learn</a> a Python package designed to machine learning in python
* <a href="https://geemap.org/" target="_blank">geemap</a> a Python package for for interactive mapping with Google Earth Engine
* <a href="https://geopandas.org/" target="_blank">geopandas</a> a Python package to make working with geospatial data in python easier. GeoPandas extends the datatypes used by pandas to allow spatial operations on geometric types.
* <a href="https://rasterio.readthedocs.io/en/latest/index.html" target="_blank">rasterio</a> a Python package to process different kind of geospatial raster data files. Rasterio reads and writes these formats and provides a Python API based on Numpy N-dimensional arrays and GeoJSON.
* <a href="https://geopython.github.io/OWSLib/" target="_blank">owslib</a> a Python package for client programming with Open Geospatial Consortium (OGC) web service (hence OWS) interface standards, and their related content models.
* <a href="https://pyepsg.readthedocs.io/en/latest/" target="_blank">pyepsg</a> a Python package for reading/writing epsg.io coordinates reference systems

## For notebook 1

```python
import cartopy.crs as ccrs
import cartopy
import geopandas
import osmnx as ox
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
import numpy as np
import cartopy.feature as cfeature
%matplotlib inline
```

```python
ccrs.epsg(32631)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})
ax.coastlines()
```

```python
import numpy as np
central_lon, central_lat = -10, 45
extent = [-20, 20, 30, 60] # with (x0, x1, y0, y1) of the map in the given coordinate system.
ax = plt.axes(projection=ccrs.Orthographic(central_lon, central_lat),title='Europe')
ax.set_extent(extent)
ax.coastlines(resolution='50m')
ax.gridlines()
scalebar = ScaleBar(1,box_alpha=0)
ax.add_artist(scalebar)
```

```python
rivers_50m = cfeature.NaturalEarthFeature('physical', 'rivers_lake_centerlines', '50m')

plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.AlbersEqualArea(central_lon, central_lat))
ax.set_extent(extent)

ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
ax.add_feature(rivers_50m, facecolor='None', edgecolor='b')
ax.gridlines()
```

```python
fig = plt.figure(figsize=(8, 12))
central_lon, central_lat = -10, 45
extent = [-20, 20, 30, 60] # with (x0, x1, y0, y1) of the map in the given coordinate system.
ax = plt.axes(projection=ccrs.Orthographic(central_lon, central_lat),title='Europe')
ax.set_extent(extent)
ax.coastlines(resolution='50m')
ax.gridlines()
ax.add_wms(wms='https://gibs.earthdata.nasa.gov/wms/epsg4326/best/wms.cgi',layers=['VIIRS_SNPP_CorrectedReflectance_TrueColor'],wms_kwargs={'time':'2021-01-01'})
```

```python
G = ox.graph_from_place('Delft, Netherlands')
ox.plot_graph(G)
```

```python
nodes, edges = ox.graph_to_gdfs(G)
nodes.head()
```

```python
filename='https://raw.githubusercontent.com/steflhermitte/EO-Notebooks/master/vegetation.geojson'
import geopandas
df = geopandas.read_file(filename)
```

```python
plt.figure()
crs=ccrs.PlateCarree()
ax = plt.axes(projection=crs)
ax.set_extent([4.36,4.39,51.98,52.01])
ax.add_geometries(edges['geometry'],crs=crs,linewidth=1, edgecolor='red',facecolor='None')
ax.add_geometries(df['geometry'],crs=crs,linewidth=0, edgecolor='None',facecolor='green')
```

## For notebook 2

```python
import geemap
import rasterio
import ee
import geemap
import os
import shutil
```

Now you need to authenticate Google Earth Engine. Therefore you first need an earth engine account for which you can register [here](https://signup.earthengine.google.com/).

```python
## Authenticate EE
ee.Authenticate()
```

```python
# Launch EE
ee.Initialize()
```

```python
roi = ee.Geometry.Polygon([[[-115.413031, 35.889467],
[-115.413031, 36.543157],
[-114.034328, 36.543157],
[-114.034328, 35.889467],
[-115.413031, 35.889467]]])

collection = ee.ImageCollection("COPERNICUS/S2_SR") \
    .filterBounds(roi) \
    .filterDate('2019-11-01', '2019-11-5') \
    .filterMetadata('CLOUDY_PIXEL_PERCENTAGE','less_than',5) \
    .select('B4') \
    .mean().clip(roi).unmask()
```

Uncomment one of the following lines to set your output directoryt (depending if work on your local computer or colab).

```python
# If on local computer
out_dir = os.path.join(os.path.expanduser('~'), 'Downloads')
# If on colab
out_dir = '/content/'
filename = os.path.join(out_dir, 'landsat.tif')
```

```python
geemap.ee_export_image(collection, filename=filename,region=roi, scale=100, file_per_band=False)
```

```python
src = rasterio.open(filename, 'r')
plt.imshow(src.read(1), cmap='pink')
```
